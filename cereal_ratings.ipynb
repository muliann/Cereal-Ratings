## Data Preprocessing

### Missing Values

```python
df.isna().sum().sum()
```

```python
(df==-1).sum().sum()
```

```python
# replaces all the -1 with nan
df=df.replace(-1,np.nan)
```

```python
# replaces all the nan values with the mean of specific column
for column in['carbo','sugars','potass']:
    df[column]=df[column].fillna(df[column].mean())
```

```python
df
```

```python
# Drop the name column since it doesn't contain any usefull information regarding the ratings
df = df.drop('name', axis=1)
df
```

### Encoding


```python
{
    column: list(df[column].unique() for column in ['mfr','type'])
}
```

```python
# Change the type values to binary encoding since they are only 2 values
for column in ['type']:
    df[column]=df[column].replace(('C',0),('H',1))
```
